```python
#Define the new x and y for the new models, dropping the cap_ features.
```

```python
yhw = dhw['spam']
Xhw = dhw.drop(columns = ['spam', 'cap_ave', 'cap_long', 'cap_total'])
```

```python
#check which columns should be binarized.
```

```python
Xhwkeep = Xhw.columns.str.startswith('word_')
print(Xhwkeep)
```

```python
# all columns that start with word_ should be binarized.
# Ignore Xhwkeep variable. 
# Binarize all Xhw columns. 
```

```python
Xhw = (Xhw>0).astype(int)
Xhw.head()
```

```python
# Set up the Decision Tree Classifier and fit it to the new data.
clhw = DecisionTreeClassifier(criterion = 'entropy', max_depth = 5)
clhw.fit(Xhw,yhw)
```

```python
round(clhw.score(Xhw,yhw),3)
```

```python
#Check with prediction and confusion matrix.
```

```python
yhw_pred = clhw.predict(Xhw)
conf_hw = confusion_matrix(yhw, yhw_pred)  
conf_hw
```

```python
tphw = conf_hw[1,1]/sum(conf_hw[1,:])
fphw = conf_hw[0,1]/np.sum(conf_hw[0,:])
round(tphw,3), round(fphw,3)
```

```python
# We see that the true positive rate of the new decision tree model is lower than 
# some of the previous models, and the false positive rate is lower. Depending on 
# the context of the problem, this may be a good tradeoff. 
```
